```python
plt.figure(1)
plt.subplot(2,2,1)
df['Gender'].value_counts().plot(figsize=(10,8),kind='bar',color='red')
plt.title("Gender Type Frequencen Diagram")
plt.ylabel("No.of Gender type")
plt.xlabel("Gender Type")



plt.subplot(2,2,2)
df['Age'].value_counts().plot(figsize=(10,8),kind='bar',color='green')
plt.title("Age Type Frequencen Diagram")
plt.ylabel("Age type ")
plt.xlabel(" Age Type")


plt.subplot(2,2,3)
df['Python'].value_counts().plot(figsize=(10,8),kind='bar',color='orange')
plt.title("Python Type Frequencen Diagram")
plt.ylabel("No.of Python type")
plt.xlabel("  Python Type")


plt.subplot(2,2,4)
df['Machine Learning'].value_counts().plot(figsize=(10,8),kind='bar',color='black')
plt.title("Machine Learning Frequencen Diagram")
plt.ylabel("No.of Machine Learning type")
plt.xlabel(" Machine Learning Type")

plt.tight_layout()
plt.show()
```
